```python
#read given csv files
train_input = pd.read_csv('/kaggle/input/home-data-for-ml-course/train.csv')
test_input = pd.read_csv('/kaggle/input/home-data-for-ml-course/test.csv')
```

```python
#from input: remove rows with Nan target
sum(train_input['SalePrice'].isnull())

#so there is no such row, nothing to remove.
```

```python
#get the Target Variable
y = train_input.SalePrice
train_input.drop(['SalePrice'], axis=1, inplace=True)

#see if output is numeric or categorical
print(y.dtype)
```

```python
#Analyze the input
train_input.head(-5)
```

```python
train_input.set_index('Id')
test_input.set_index('Id')
```

```python
#get object_cols, numeric_cols
object_cols = [col for col in train_input.columns if train_input[col].dtype =='object']
numeric_cols = [col for col in train_input.columns if train_input[col].dtype !='object']

print(len(train_input.columns))
print(len(object_cols))
print(len(numeric_cols))
```

```python
#get good_obj_cols, bad_obj_cols
good_obj_cols = [col for col in object_cols if 
                    set(test_input[col]).issubset(set(train_input[col]))]

bad_obj_cols = list(set(object_cols)-set(good_obj_cols))

print(len(good_obj_cols))
print(len(bad_obj_cols))
```

```python
#drop bad_obj_cols from both
X_train = train_input.drop(bad_obj_cols,axis=1)
X_valid = test_input.drop(bad_obj_cols,axis=1)
```

```python
len(X_train.columns)
```

```python
#for extended impute:
for col in X_train.columns:
    if(X_train[col].isnull().any()):
        X_train[col+'_missing']=X_train[col].isnull()
        X_valid[col+'_missing']=X_valid[col].isnull()
```

```python
len(X_train.columns)
```

```python
#get columns
categorical_cols = [col for col in X_train.columns if X_train[col].dtype == 'object']
numerical_cols = [col for col in X_train.columns if X_train[col].dtype != 'object']
```

```python
#Imputers:
from sklearn.impute import SimpleImputer

#Numerical Transformer
numerical_transformer = SimpleImputer(strategy='median')
```

```python
high_cols = [col for col in categorical_cols if X_train[col].nunique() > 10]
low_cols = [col for col in categorical_cols if X_train[col].nunique() <= 10]
```

```python
from sklearn.pipeline import Pipeline, FeatureUnion
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OrdinalEncoder, OneHotEncoder

categorical_high_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown='ignore',sparse=False))
])

categorical_low_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('ordinal', OrdinalEncoder())
])
```

```python
#PreProcessor
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numerical_transformer, numerical_cols),
        ('cat1', categorical_high_transformer, high_cols),
        ('cat2', categorical_low_transformer, low_cols)
    ])
```

```python
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor(n_estimators=100, random_state=0)
```

```python
# Bundle preprocessing and modeling code in a pipeline
my_pipeline = Pipeline(steps=[('preprocessor', preprocessor),
                              ('model', model)
                             ])
```

```python
my_pipeline.fit(X_train, y)
```

```python
preds_test = my_pipeline.predict(X_valid)
output = pd.DataFrame({'Id': X_valid.index,
                       'SalePrice': preds_test})
output.to_csv('submission.csv', index=False)
```
